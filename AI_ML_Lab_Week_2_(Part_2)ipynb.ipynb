```
data = pd.read_csv("/content/multiple_linear_data.csv", header=0)
data = data[(data['G3']!=0)|((data['G2']==0)&(data['G3']==0))]
#data.insert(0,"bias", 1)

data[:5]
```

```
# You may need original list of columns to interpret the features after transformation, maybe
orig_cols = data.columns
```

```
# Handle categorical values
from sklearn.preprocessing import OneHotEncoder,LabelEncoder
data = data.dropna()
#enc=OneHotEncoder(handle_unknown= 'ignore')
##enc.fit_transform(data[['address',	'famsize',	'reason',	'studytime',	'failures',	'schoolsup',	'famsup',	'paid',	'activities',	'higher',	'internet',	'romantic',	'freetime',	'goout',	'health']]).toarray()
#data = enc.fit_transform(data[['address',	'famsize',	'reason',	'schoolsup',	'famsup',	'paid',	'activities',	'higher',	'internet',	'romantic']]).toarray()
cat=['address',	'famsize',	'reason',	'schoolsup',	'famsup',	'paid',	'activities',	'higher',	'internet',	'romantic']
le = LabelEncoder()
for i in cat:
  le.fit(data[i])
  data[i]= le.transform(data[i])

#enc.categories_
data
```

```
X= data.drop(['G3'], axis=1)
y= data['G3']
```

```
reg = LinearRegression().fit(X, y)
reg.score(X,y)
```

```
y_pred= reg.predict(X)

from sklearn import metrics
print("Mean squared error:",metrics.mean_squared_error(y,y_pred))
```
